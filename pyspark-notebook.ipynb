```python
# Challenge: Essential data manipulation

# 1. Read data (Jan and Feb 2025)
df_jan_2025 = spark.read.parquet(r"dataset\yellow_tripdata_2025-01.parquet")
df_feb_2025 = spark.read.parquet(r"dataset\yellow_tripdata_2025-02.parquet")
```

```python
# 2. Compute union of the two dataframes
df_2025_combined = df_jan_2025.union(df_feb_2025)
```

```python
# 3. Select and rename columns
df_2025_combined = df_2025_combined.\
select('tpep_pickup_datetime', 'tpep_dropoff_datetime',\
       'PULocationID', 'DOLocationID', 'passenger_count',\
       'fare_amount', 'Airport_fee', 'total_amount', 'payment_type').\
withColumnsRenamed({
    'tpep_pickup_datetime': 'pu_datetime',\
    'tpep_dropoff_datetime': 'do_datetime',\
    'PULocationID': 'pu_location_id',\
    'DOLocationID': 'do_location_id',\
    'Airport_fee' : 'airport_fee'})
```

```python
# 4. Read Taxi_zone csv file 
taxi_zone_lookup = spark.read.option('header','true').csv(r"dataset\taxi_zone_lookup.csv")
```

```python
# 5. Join Taxi_zone_lookup with df_2025_combined 
df_2025_combined = df_2025_combined.join(taxi_zone_lookup, df_2025_combined.do_location_id == taxi_zone_lookup.LocationID, 'left')
```

```python
# 6. Drop and rename columns
df_2025_combined = df_2025_combined.drop('LocationID', 'Zone', 'service_zone').withColumnsRenamed({'Borough': 'pu_boro'})
```

```python
# Challenge : PySpark SQL

# 1. Read January Taxi Ride data and register a temporary view
taxi_jan2025 = spark.read.parquet(r"dataset\yellow_tripdata_2025-01.parquet")
taxi_jan2025.createOrReplaceTempView('taxi_jan2025')

# 2. Read taxi lookup csv and register a temporary view
taxi_zone_lookup = spark.read.option('header','true').csv(r"dataset\taxi_zone_lookup.csv")
taxi_zone_lookup.createOrReplaceTempView('taxi_zone_lookup')
```

```python
taxi_jan2025.printSchema()
taxi_zone_lookup.printSchema()
```

```python
# 3. Join the two views based on locationID
challenge_query = '''
SELECT DOLocationID, Borough, total_amount
FROM taxi_jan2025 LEFT JOIN taxi_zone_lookup
ON DOLocationID = LocationID
'''

joined_df = spark.sql(challenge_query)
joined_df.show()
```

```python
# 4. Compute average with pyspark df by grouping borough column taking average on total_amount
result_df = joined_df.groupby('Borough').agg(avg('total_amount').alias('avg_amount'))
result_df.show()
```
